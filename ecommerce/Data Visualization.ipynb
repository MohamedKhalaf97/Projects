```python
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
sns.set(rc={'figure.figsize': [8, 8]}, font_scale=1.2)
```

```python
df = pd.read_csv('Data/Ecommerce.csv')
df
```

```python
sns.distplot(df['Discount_offered'])
```

```python
sns.distplot(df['Cost_of_the_Product'])
```

```python
sns.jointplot(x='Cost_of_the_Product', y='Discount_offered', data=df, kind='reg', height=8)
```

```python
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.jointplot(x='Cost_of_the_Product', y='Discount_offered', data=df, kind='kde', cmap=cmap, height=8)
sns.jointplot(x='Cost_of_the_Product', y='Discount_offered', data=df, kind='hex', cmap=cmap, height=8)
```

```python
sns.pairplot(df, vars=['Cost_of_the_Product','Discount_offered'], height=5, aspect=1.3)
```

```python
sns.boxplot(x="Gender", y="Discount_offered", data=df)
```

```python
sns.boxplot(x="Gender", y="Cost_of_the_Product", data=df)
```

```python
sns.boxplot(x="Gender", y="Cost_of_the_Product", hue="Mode_of_Shipment",data=df)
```

```python
sns.boxplot(x="Gender", y="Discount_offered", hue="Mode_of_Shipment",data=df)
```

```python
sns.boxplot(x="Gender", y="Customer_care_calls", hue="Mode_of_Shipment",data=df)
```

```python
sns.boxplot(x="Gender", y="Customer_care_calls", hue="Warehouse_block",data=df)
```

```python
sns.boxplot(x="Gender", y="Customer_rating", hue="Warehouse_block",data=df)
```

```python
sns.violinplot(x="Mode_of_Shipment", y="Cost_of_the_Product", data=df, hue='Gender', palette='viridis')
```

```python
sns.stripplot(x="Mode_of_Shipment", y="Cost_of_the_Product", data=df, hue='Gender', dodge=True)
```

```python
sns.boxplot(x="Product_importance", y="Cost_of_the_Product", data=df)
sns.stripplot(x="Product_importance", y="Cost_of_the_Product", color="#474646", size=5, data=df)
```

```python
sns.violinplot(x="Product_importance", y="Discount_offered", data=df)
sns.stripplot(x="Product_importance", y="Discount_offered", color="#474646", size=5, data=df)
```

```python
Values = df.pivot_table(values='Discount_offered', index='Customer_care_calls', columns='Customer_rating')
Values
```

```python
sns.heatmap(Values, linecolor='#FFFFFF', linewidths=4)
```

```python
cmap = sns.cubehelix_palette(as_cmap=True, reverse=True, dark=0, light=1)
sns.heatmap(Values, linecolor='#FFFFFF', linewidths=2, cmap=cmap)
```
